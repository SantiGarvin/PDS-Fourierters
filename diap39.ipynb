```python
import numpy as np
from IPython.display import Audio

f_S = 8000  # Frecuencia de muestreo (Hz)
f_0 = 2000   # Frecuencia del tono (Hz)
t = 2        # Duraci√≥n en segundos

# Generar el array de tiempo
time = np.linspace(0, t, int(f_S * t), endpoint=False)  # Vector de tiempo

# Crear la onda sinusoidal
tone = np.sin(2 * np.pi * f_0 * time)  

# Reproducir el audio
Audio(tone, rate=f_S)
```

```python
n = np.random.randn(tone.size)

Audio(n, rate=f_S)
```

```python
y = n + tone

Audio(y, rate=f_S)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 4))
plt.plot(time[:100], tone[:100])  
plt.title("Se√±al de Tono de 2000 Hz")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.grid()
plt.show()
```

```python
Y = np.fft.rfft(y)
plt.plot(np.abs(Y))
plt.show()
print(Y.size)
```

```python
X = np.fft.fft(y) # te decvuleve los coeficientes de la transformada simetrica
                 # con un vecotr con las mismas muestras (16000 muestras ya que era 8000)
                 # rfft te devuelve la parte (8000) ya que con eso nos basta
print(X.size)
plt.plot(np.abs(X))
plt.show()
```

```python
freqs=np.fft.rfftfreq(y.size)

plt.plot(freqs,np.abs(Y))
plt.xlabel("Frecuencia")
plt.ylabel("Modulo FFT")

```

```python
# asi estaria normalizaad la frecuencia f=f0/FS que nomralizada es K/N 
# pues FS*K/N para desnomralizar es dece
freqs=f_S*np.fft.rfftfreq(y.size)

plt.plot(freqs,np.abs(Y))
plt.xlabel("Frecuencia")
plt.ylabel("Modulo FFT")

# primer valor simpre 0 y el ultmo la mitad de la frecuencia de muestreo
```

```python
# para indentificar los picos (muy importante en el decode)

ind_freq_max = np.argmax(np.abs(Y))
freq_max = freqs[ind_freq_max]
print (freq_max)


# para la practica
#  podemos dividir el array de 1000 para abajo yde mil para arriba hacemos 2 veces esto
# para obtener freqa bahos y altos
# utilizar argsort de np para que sean los primeros valors de los 2 arrays que obtenemos 

# una vez obtenidos los picos ver a que frecuencia esta mas cerca de las de la tabla
```
